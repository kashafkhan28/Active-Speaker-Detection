```python
def speaker_detection(fused_score, img1_faces):
  threshold = 0.5
  max_score=max(fused_score)
  score_val=fused_score.index(max_score)
  facial_cord=[]
  if max_score > threshold:
    for value in img1_faces.values():
      facial_cord.append(value['facial_area'])
    active_speaker = facial_cord[score_val]
  else:
    active_speaker = 'None'
  return active_speaker
```
