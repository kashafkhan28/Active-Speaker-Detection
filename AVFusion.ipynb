```python
# Sample normalized prediction scores from audio and video models

def av_fusion(persons, audio_score):

  fused_scores=[]
  for i in range(0,len(persons)):
    score = audio_score+persons[i]
    fused_scores.append(score)
  return fused_scores
```

```python
def find_key_by_value(dictionary, target_value):
    for key, value in dictionary.items():
        if value == target_value:
            return key
```
